# Question 1

### How does knowing data type (qualitative vs quantiative) provide insights in the type of distribution a random variable comes from?
Typically qualitative data can only be measured in terms of discrete variables, as there are finite number of variables that the observations can take on. In contrast, quantiative data can be either discrete or continuous (meaning it can take on any possible value within a given range). Furthermore, the number of values that a discrete qualitative variable can take will also influence the type of distribution that a variables comes from.

For example, the Bernoulli distribution is useful when there are 2 possible discrete outcomes (e.g., flipping a coin), whereas the uniform distrubtion more closely describes multiple discrete outcomes (e.g., a flat probability distribution function, which you would expect to see for fair 6-sided dice roll). If the data are quantitative, then it tells you that it might follow a different distribution, such as a normal distribution. This is because there are infinite number of possible outcomes in between whole integers, so the distribution looks like a bell-shaped curve.

This is a uniform distribution, and therefore the mean is given by the equation:

$\mu = \frac{1}{2} (1 + 6)$ = 3.5

The standard deviation is given by:

$\sigma^2 = \sum_{i = 1}^{6} (i - \mu)^2 f(i) $

where i is the value of a given die face, $\mu$ is the mean (in this case, 3.5), and f(i) is the frequency at which it is rolled (1/6 for a fair die).

Another way to think about the standard deviation is to think about it in terms of a classic variance equation, such that you take each possible value and subtract the mean (3.5), square that term, and multiply it by the probability of getting that value (1/6, in this case). See below for calculation.

```r
# Mean of fair dice roll
0.5 * (1+6)
```

```r
# Standard deviation of fair dice roll, uniform distribution.
sqrt((6.25 + 2.25 + 0.25 + 0.25 + 2.25 + 6.25) * (1/6))
```

# Question 4 
#### The goal here is to set up an experiment where we can adjust the number of throws and the number of dice thrown. 
We want to plot the sums of each of the throws. We can do this by setting it up as a matrix where each column is a single throw and each row represents the number of die thrown. By summing the columns together, we can find the sum of each set of throws.

```r
# set up a single die rolled as a function where it is thrown n times.
die = function(n) {
    return(as.integer(runif(n, 1, 6+1)))
}
die(10) # roll a single die 10 times
```

```r
#set it up as a matrix, example here is 10 throws of 2 dice each.
exmat = matrix(die(20), 2, 10)
exmat
```

Having defined our 'die' function, going to set this up into a function so we can easily adjust throws and dice number.

```r
dice = function(T = 10, N = 2) {
    m = matrix(die(N * T), N, T)
    return(colSums(m))}
```

```r
dice() #rolls the default 2 dice 10 times. Should result in 10 values in a single row, since the return is colSum
```

```r
dice(10, 3) #this rolls 3 dice 10 times. Should result in 10 values.
```

```r
Exp1 = dice(10000, 1)
hist(Exp1, main = ("1 Die"), breaks = seq(0.5, 6.5, 1), xlab = ("Value on Top Die Face"), ylim = c(0,2000))
```

```r
Exp2 = dice (10000, 2)
hist(Exp2, main = ("2 Dice"), xlab = ("Sum of 2 Dice"), ylim = c(0,2000))
```

```r
Exp3 = dice (10000, 3)
hist(Exp3, main = ("3 Dice"), xlab = ("Sum of 3 Dice"), ylim = c(0,1500))
```

```r
Exp4 = dice (10000, 6)
hist(Exp4, main = ("6 Dice"), xlab = ("Sum of 6 Dice"), ylim = c(0,2000))
```

Based on these graphs, it seems that the graph throwing 6 dice 10,000 times produces the distribution that most closely resembles a normal distribution.

```r
qqnorm(Exp1, pch = 1, frame = FALSE, ylim = c(0, 8))
qqline(Exp1, col = "steelblue", lwd = 2)
```

```r
qqnorm(Exp2, pch = 1, frame = FALSE, ylim = c(0, 15))
qqline(Exp2, col = "steelblue", lwd = 2)
```

```r
qqnorm(Exp3, pch = 1, frame = FALSE, ylim = c(0, 25))
qqline(Exp3, col = "steelblue", lwd = 2)
```

```r
qqnorm(Exp4, pch = 1, frame = FALSE, ylim = c(0, 45))
qqline(Exp4, col = "steelblue", lwd = 2)
```

### Conclusions:
From these qq plots, it is clear that Exp 4 (rolling 6 dice 10,000 times) produces data that most closely approximates the line (in blue) that represents a perfectly normal distribution. This is consistent with the histogram.
