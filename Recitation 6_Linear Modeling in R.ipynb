```r
#attach allows you to avoid writing as much; if you attach several different datasets. If you have 1 dataframe, then you're good to go
library(MASS)
library(ISLR)
attach(Boston)
options(repr.plot.width=4, repr.plot.height=4)
plot(lstat, medv, xlab="Lower status of population (percnet)", ylab="Median house value")
lm.fit = lm(medv~lstat)
summary(lm.fit)
```

```r
#This grabs the coefficients themselves.
coef(lm.fit)
```

```r
# This is how we get the
confint(lm.fit)
```

```r
# The . in this function tells R to use all the variables in the model (full factorial)
lm.fit=lm(medv~., data=Boston)
summary(lm.fit)
```

So, from the above, it is clear that age does not have an impact on the overall model. So how can we remove it from the overall model?

```r
# Excluding a variable from the model: age, indus
lm.fit_new = lm(medv~.-age -indus, data=Boston)
summary(lm.fit_new)
```
